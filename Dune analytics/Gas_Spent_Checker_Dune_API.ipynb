```python
from dune_client.types import QueryParameter
from dune_client.client import DuneClient
from dune_client.query import QueryBase
```

### Input our Dune API key here and query ID here üëá

```python
dune_api_key = '<paste your api key here>' # input your API key, to create one follow guide here https://docs.dune.com/api-reference/overview/authentication
```

```python
dune = DuneClient(
    api_key=dune_api_key,
    base_url="https://api.dune.com",
    request_timeout=(300) # request will time out after 300 seconds
)
```

## Get latest result for [Gas Spent Checker)](https://dune.com/queries/2860880)




```python
# Define query parameters
start_date = "2024-01-01 00:00:00"
end_date = "2024-06-01 00:00:00"
wallet_address = "0xAda62a424E8672634052DaFE4ecAe86cdf9a23B7"
```

```python
query = QueryBase(
    query_id=2860880,
    params=[
        QueryParameter.date_type(name="Start Date", value=start_date),
        QueryParameter.date_type(name="End Date", value=end_date),
        QueryParameter.text_type(name="Wallet Address", value=wallet_address),
    ],
)
```

```python
query_result = dune.run_query_dataframe(query=query)
```
